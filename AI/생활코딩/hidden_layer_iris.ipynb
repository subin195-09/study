```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 1. 과거의 데이버 준비 (iris)
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(path)

# one-hot encoding
iris = pd.get_dummies(iris)

#독립, 종속 분리
독립 = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(독립.shape, 종속.shape)
```

```
# 2. 모델의 구조 만들기

X = tf.keras.layers.Input(shape = [4])
H = tf.keras.layers.Dense(5, activation = 'swish')(X)
H = tf.keras.layers.Dense(5, activation = 'swish')(X)
H = tf.keras.layers.Dense(5, activation = 'swish')(X)
Y = tf.keras.layers.Dense(3, activation = 'softmax')(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy')
```

```
# 3. 모델 학습(fit)
model.fit(독립, 종속, epochs=1000, verbose=0)
model.fit(독립, 종속, epochs=10)
```

```
# 4. 모델 이용
print(model.predict(독립[0:5]))
print(종속[0:5])
```
