```
# 라이브러리
import pandas as pd
```

```
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
iris = pd.read_csv(path)
iris.head()
```

```
# NA값을 체크
iris.isna().sum()
```

```
# 맨 마지막 값이 빈값 (NaN = not available) 이기 때문에 에러가 난다
# NA 값을 수정해주어야 한다
iris.tail()
```

```
# 해결 : NA값에 꽃이폭 평균값을 넣어준다
mean = iris['꽃잎폭'].mean()
print(mean)
iris['꽃잎폭'] = iris['꽃잎폭'].fillna(mean)
iris.tail()
```
