```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 1. 과거의 데이터 준비

path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(path)

iris.head()
```

```
# one hot encoding
encoding = pd.get_dummies(iris)
encoding.head()
```

```
print(encoding.columns)
```

```
# 독립, 종속 분류

독립 = encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = encoding[['품종_setosa', '품종_versicolor', '품종_virginica']]

print(독립.shape, 종속.shape)
```

```
# 2. 모델의 구조를 만든다

X = tf.keras.layers.Input(shape = [4])
Y = tf.keras.layers.Dense(3, activation = 'softmax')(X)

model = tf.keras.models.Model(X, Y)
model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy')
```

```
# 3. 모델 학습(fit)
model.fit(독립, 종속, epochs=10000, verbose = 0)
model.fit(독립, 종속, epochs=10)
```

```
# 4. 모델 이용
model.predict(독립[-5:])
```

```
print(종속[-5:])
```

```
# 학습한 가중치 출력
model.get_weights()

# [setosa에 대한 가중치, versicolor에 대한 가중치, virginica에 대한 가중치]
# [setosa에 대한 bias, versicolor에 대한 bias, virginica에 대한 bias]
```
