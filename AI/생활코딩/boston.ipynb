```
# 라이브러리 import하기
import tensorflow as tf
import pandas as pd
```

```
# 1. 과거의 데이터 준비
#path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
path = '/content/boston.csv'
boston = pd.read_csv(path)
boston.head()
print(boston.columns)
```

```
# 1 - 2. 독립변수, 종속변수 분리
독립 = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
종속 = boston[['medv']]

print(독립.shape, 종속.shape)
```

```
# 2. 모델 구조 만들기
X = tf.keras.layers.Input(shape = [13])
Y = tf.keras.layers.Dense(1)(X)

model = tf.keras.models.Model(X, Y)
model.compile(loss = 'mse')
```

```
# 3. 데이터로 모델을 학습(fit)시킨다.
model.fit(독립, 종속, epochs = 10000, verbose= 0)
model.fit(독립, 종속, epochs = 10)
```

```
# 4. 모델 이용해 보기
#독립 변수 첫번째 부터 5번째 까지 slicing해서 가지고 오기
model.predict(독립[0 : 5])
```
