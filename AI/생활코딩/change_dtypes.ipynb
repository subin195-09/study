```
# 라이브러리
import pandas as pd
```

```
# 기존 iris 데이터와 달리 품종을 숫자로 표현
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
iris = pd.read_csv(path)
iris.head()
```

```
# one hot encoding
# 품종이 숫자 이기 때문에 범주형으로 인식을 못해 원핫인코딩이 되지 않는다
encoding = pd.get_dummies(iris)
encoding.head()
```

```
print(iris.dtypes)
```

```
# 품종의 type을 범주형으로 바꾸어 준다
iris['품종'] = iris['품종'].astype('category')
print(iris.dtypes)
```
